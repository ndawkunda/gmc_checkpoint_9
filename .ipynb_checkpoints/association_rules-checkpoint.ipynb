```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules 
```

# DATASET 1

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
    ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

df_1 = pd.DataFrame(dataset)
df_1
```

```python
df_res_1 = pd.DataFrame()
for i in range(len(df_1.columns)):
    df_res_1 = df_res_1.append(df_1[i].value_counts())

df_res_1
```

```python
df_sum_1 = df_res_1.sum().sort_values(ascending=False)
df_sum_1
```

### Plotting

```python
plt.figure(figsize=(14,8))
plt.title("FREQUENCY")
color = plt.cm.viridis(np.linspace(0, 1, 40))
df_sum_1.plot.bar(color = color)
plt.xticks(rotation = 'vertical')
plt.show()
```

### Apriori algorithm

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_1 = pd.DataFrame(te_ary, columns=te.columns_)
df_1
```

```python
frequent_itemsets_df_1 = apriori(df_1, min_support=0.6, use_colnames=True)
frequent_itemsets_df_1
```

```python
association_rules(frequent_itemsets_df_1, metric="confidence", min_threshold=0.7)
```

```python
association_rules(frequent_itemsets_df_1, metric="lift", min_threshold=1.25)
```

# DATASET 2

```python
df_2 = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
df_2
```

```python
df_res_2 = pd.DataFrame()
for i in range(len(df_2.columns)):
    df_res_2 = df_res_2.append(df_2[i].value_counts())

df_res_2.head()
```

```python
df_sum_2 = df_res_2.sum().sort_values(ascending=False)
df_sum_2
```

### Plotting

```python
plt.figure(figsize=(14,8))
plt.title("FREQUENCY")
color = plt.cm.viridis(np.linspace(0, 1, 40))
df_sum_2.head(50).plot.bar(color = color)
plt.xticks(rotation = 'vertical')
plt.show()
```

### Apriori algorithm

```python
arr = []
for i in range(df_2.shape[0]):
    arr.append([str(df_2.values[i,j]) for j in range(df_2.shape[1])])

arr = np.array(arr)

te_ary = te.fit(arr).transform(arr)
df_2 = pd.DataFrame(te_ary, columns=te.columns_)
pd.set_option('display.max_columns', None)
df_2
```

```python
df_2 = df_2.drop(["nan"], axis=1)
df_2
```

```python
frequent_itemsets_df_2 = apriori(df_2, min_support=0.01, use_colnames=True)
frequent_itemsets_df_2
```

```python
association_rules(frequent_itemsets_df_2, metric="confidence", min_threshold=0.4)
```

```python
association_rules(frequent_itemsets_df_2, metric="lift", min_threshold=2.4)
```
